```python
# Se importa las librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Se carga el dataset Iris
df = sns.load_dataset('iris')
```

```python
# Exploración inicial de los datos
df.head()       # Primeras filas
df.info()       # Información general del dataset
df.describe()   # Estadísticas básicas
```

```python
# Análisis de valores faltantes
df.isnull().sum()
```

```python
# Visualización – Histogramas
df.hist(figsize=(10, 8))
plt.suptitle('Histogramas de variables numéricas')
plt.show()
```

```python
# Mapa de calor de correlaciones
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Mapa de calor de correlaciones')
plt.show()
```

```python
# Detección de patrones y anomalías (boxplot)
sns.boxplot(data=df, orient='h')
plt.title('Boxplot - Outliers')
plt.show()
```

```python
# Conclusiones
print("1. No se encontraron valores nulos.")
print("2. Se observa alta correlación entre petal_length y petal_width.")
print("3. Existen posibles outliers en sepal_width.")
```
